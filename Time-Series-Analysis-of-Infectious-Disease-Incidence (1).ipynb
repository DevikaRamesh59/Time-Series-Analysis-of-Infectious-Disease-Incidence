## **Outlier Treatment**

```python


Q1 = df['OBS_VALUE'].quantile(0.25)
Q3 = df['OBS_VALUE'].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Normal values range: {lower_bound} to {upper_bound}")

# Outliers
outliers = df[(df['OBS_VALUE'] < lower_bound) | (df['OBS_VALUE'] > upper_bound)]
print(f"Total outliers: {len(outliers)}")
```

### **2. Histogram (Numerical Features)**
Histogram mainly focusus on Continuous/Numerical Data (Incidence rates, Age, Years).

**Central Tendency:** It helps you visually estimate where the "center" of the data lies (Mean/Median).

**Spread and Range:** We can see the minimum and maximum values and how the data is spread across that range.

**Skewness:** It reveals if the data is symmetric (Bell curve) or skewed to the right/left (which we checked earlier).

**Kurtosis (Peakedness):** It shows how sharp or flat the peak of the data distribution is.

### **3. Boxplot (Outlier Detection)**

```python
### Boxplot for Outliers
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['OBS_VALUE'])
plt.title('Boxplot of Observation Values')
plt.savefig('obs_value_boxplot.png')
```

### **3.Boxplot (Statistical Summary):**
A Boxplot is a powerful tool used to visualize the distribution of numerical data through their quartiles. It is fundamentally based on the "5-Number Summary," which provides a quick statistical overview of the dataset.

**The 5-Number Summary:**
A Boxplot visually represents these five specific statistical values:

**Minimum (Q0):** The lowest data point excluding any outliers.

**First Quartile (Q1 / 25th Percentile):** The middle value between the minimum and the median. 25% of the data falls below this point.

**Median (Q2 / 50th Percentile):** The middle value of the dataset. It shows the center of the data.

**Third Quartile (Q3 / 75th Percentile):** The middle value between the median and the maximum. 75% of the data falls below this point.

**Maximum (Q4):** The highest data point excluding any outliers.

### **3.Correlation Heatmap (Multiple Numerical Variables)**

```python
### Correlation matrix
corr = df[['TIME_PERIOD', 'OBS_VALUE']].corr()

### Heatmap plot
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```

 **3.Correlation Heatmap (Multiple Numerical Variables)**

**Features:**

**Correlation Coefficient:** Displays values between -1 to +1.

**+1:** Perfect positive correlation.

**0:** No relationship.

**-1:** Perfect negative correlation.

**Feature Selection:** Helps identify redundant variables (highly correlated ones) that can be removed.

**Color Encoding:** Uses a color scale (e.g., Heat/Cool) to quickly spot the strongest relationships.

### **Multivariate Analysis → relation among 3+ variables (pairplot, grouped analysis, pivot tables, advanced plots)**

### **1. Pairplot (Variable Relationships)**

```python
sns.pairplot(df[['TIME_PERIOD','OBS_VALUE','Disease']], hue ='Disease')
plt.show()
```

**Pair Plot Features:**

**Matrix Visualization:** Creates a grid of scatter plots for every pair of numerical features.

**Distribution View:** The diagonal shows the univariate distribution (histogram/KDE) of each variable.

**Pattern Recognition:** Helps identify how different groups (e.g., different diseases) behave differently across all variables.

### **2.Pivot Tables (Data Summarization)**

```python
pivot_table = pd.pivot_table(df,
                       values='OBS_VALUE',
                       index='Reference area',
                       columns='Disease',
                       aggfunc='mean')
pivot_table
```

**Pivot tabele Features:**

**Multi-Dimensional Summary:** Summarizes data across two or more dimensions (e.g., Rows = Countries, Columns = Diseases).

**Aggregation:** Allows you to apply different mathematical functions like mean, sum, or count simultaneously.

**Cross-Tabulation:** Helps in comparing how different categories interact with each other numerically.

### **Group by Summary**

```python
# Country and Disease wise summary
summary = df.groupby(['Reference area', 'Disease'])['OBS_VALUE'].agg(['mean', 'max', 'count'])
print(summary.sort_values(by='mean', ascending=False))
```

**Advanced Visualization (Heatmap from Pivot Table)**

```python
## Plot the Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table,
            annot=True,
            cmap='YlGnBu')

plt.title('Disease Incidence Heatmap')
plt.show()
```

**Heatmap from Pivot Table Features:**
A Heatmap transforms a complex Pivot Table into an intuitive color-coded matrix.


**Color Encoding:** Uses a color gradient (e.g., Light to Dark) to represent numerical values. Darker or more intense colors usually represent higher values (Hotspots).

**Pattern Recognition:** It allows you to instantly spot clusters or "hot zones" where disease incidence is highest across different countries and diseases.

**Annotated Matrix:** By setting annot=True, the actual numerical values are displayed inside the cells, combining the precision of a table with the speed of a chart.

**Data Density:** It effectively handles a large number of categories (like 20 countries vs 10 diseases) without making the chart look cluttered or messy.

**Comparative Analysis:** You can easily compare vertically (which country has the most issues?) and horizontally (which disease is the most widespread?).

### **Multivariate Scatter Plot**

```python


# 1. Plotting with Hue and Style
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df,
                x='TIME_PERIOD',
                y='OBS_VALUE',
                hue='Disease',
                style='Reference area',
                size='OBS_VALUE',
                alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.title('Multivariate Analysis: Year, Incidence, Disease, and Country')
plt.show()
```

**Multivariate Scatter Plot Features:**

A Multivariate Scatter Plot allows you to visualize more than two variables in a two-dimensional space by using visual cues like color, size, and shape.

**Dimensionality Expansion:** It adds depth to a 2D plot. Instead of just seeing Year vs. Value, you now see how different diseases behave across different countries over time.

**Pattern Comparison:** You can easily compare the "Growth Trajectory" of one disease in Country A versus Country B.

**Trend Isolation:** Even if two diseases have similar incidence values, different colors and shapes help you isolate and track their individual trends.

**Visual Grouping:** It reveals "Hidden Segments" – for example, you might notice that a specific disease is consistently higher in one country compared to others across all years.
